```
#!pip install flatten-dict
import requests
from requests_oauthlib import OAuth1
import json
import pandas as pd
from pandas.io.json import json_normalize
from flatten_dict import flatten

api_url_base = 'https://api.propublica.org/congress/v1/statements/search.json?query=comey&offset=0'
headers = {'X-API-Key': 'JNCyGEYgLzSdaCqfQXataeL8i0Gx4E1R7AJHPUwn'}


#Senate
Senate_116= 'https://api.propublica.org/congress/v1/116/senate/members.json?offset=0'
r_116S = requests.get(Senate_116, headers=headers )
DataSenateMembers=r_116S.json()

df_json=json_normalize(DataSenateMembers['results'])
df_members=df_json['members'][0]
df_SenMembers=pd.DataFrame(df_members)

#House
House_116= 'https://api.propublica.org/congress/v1/116/house/members.json?offset=0'
r_116H = requests.get(House_116, headers=headers )
DataHouseMembers=r_116H.json()

df_jsonH=json_normalize(DataHouseMembers['results'])
df_Hmembers=df_jsonH['members'][0]
df_HouseMembers=pd.DataFrame(df_Hmembers)

```

```
#Bills

#Introduced
IntroBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/introduced.json'
r_116IntroB = requests.get(IntroBills_116, headers=headers)
IntroBills=r_116IntroB.json()

df_jsonIntroB=json_normalize(IntroBills['results'])
df_Introbills=df_jsonIntroB['bills'][0]
Introduced=pd.DataFrame(df_Introbills)

#updated

UpdatedBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/updated.json'
r_116UpB = requests.get(UpdatedBills_116, headers=headers)
UpBills=r_116UpB.json()

df_jsonUpB=json_normalize(UpBills['results'])
df_Upbills=df_jsonUpB['bills'][0]
Updated=pd.DataFrame(df_Upbills)

#active
ActBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/active.json'
r_116ActB = requests.get(ActBills_116, headers=headers)
ActBills=r_116ActB.json()

df_jsonActB=json_normalize(ActBills['results'])
df_Actbills=df_jsonActB['bills'][0]
Active=pd.DataFrame(df_Actbills)

#passed
PassedBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/passed.json'
r_116PB = requests.get(PassedBills_116, headers=headers)
PBills=r_116PB.json()

df_jsonPB=json_normalize(PBills['results'])
df_Pbills=df_jsonPB['bills'][0]
Passed=pd.DataFrame(df_Pbills)

#enacted
enactedBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/enacted.json'
r_116enactedB = requests.get(enactedBills_116, headers=headers)
enactedBills=r_116enactedB.json()

df_jsonEB=json_normalize(enactedBills['results'])
df_Ebills=df_jsonEB['bills'][0]
Enacted=pd.DataFrame(df_Ebills)

#vetoed
vetoedBills_116= 'https://api.propublica.org/congress/v1/116/both/bills/vetoed.json'
r_116vetoedB = requests.get(vetoedBills_116, headers=headers)
VBills=r_116vetoedB.json()

df_jsonVB=json_normalize(VBills['results'])
df_Vbills=df_jsonVB['bills'][0]
Vetoed=pd.DataFrame(df_Vbills)
```
